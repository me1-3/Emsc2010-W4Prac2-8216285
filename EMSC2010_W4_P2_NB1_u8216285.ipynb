## Plotting the Keeling curve

This record is referred to as the *Keeling curve* in honour of Charles Keeling who started the monitoring program and ran it until his death in 2005.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Data:**

*Globally averaged marine surface monthly mean data:*

* Lan, X., Tans, P. and K.W. Thoning: Trends in globally-averaged CO2 determined from NOAA Global Monitoring Laboratory measurements. Version 2026-03 https://doi.org/10.15138/9N0H-ZH07

```python
spreadsheet_name = 'co2_mm_mlo.xlsx'
worksheet_name = 'co2_mm_mlo' #name of your worksheet containing the data
df = pd.read_excel(spreadsheet_name, sheet_name=worksheet_name) #read the data into a dataframe
df.head()
```

```python
#generating the plot
fig, ax = plt.subplots(figsize=(12,5)) #set figure size to 15 inches by 10 inches
ax.plot( #plot time vs. average CO2 ppm
    df['decimal date'],
    df['average'],
    'r', #make line red
    alpha=0.5, #make line semi transparent
    label='Monthly Average')
ax.plot( #plot time vs. deseasonalized CO2 ppm on ax
    df['decimal date'],
    df['deseasonalized'],
    'k', #colour black
    label='Deseasonalized')
ax.set_xlabel(
    'Year',
    fontstyle='italic') #labeling my x-axis & make it italic
ax.set_ylabel(
    r'$\text{CO}_2$ concentration (ppm)',
    fontstyle='italic') #labeling my y-axis & make it italic
ax.set_title(
    r'Global average $\mathbf{CO_2}$ concentration',
    fontweight='bold') #labeling my title, adding subscript & make bold
ax.minorticks_on() #show the minor ticks
ax.legend(loc='upper left', labelcolor = 'linecolor') #show the legend with the labels given above in upper left corner
plt.grid(True)
```

```python
#zooming in on the plot
fig, ax = plt.subplots(figsize=(12,5)) #set figure size to 15 inches by 10 inches
ax.plot( #plot time vs. average CO2 ppm on ax
    df['decimal date'],
    df['average'],
    'r', #make line red
    label='Average')
ax.plot( #plot time vs. deseasonalized CO2 ppm on ax
    df['decimal date'],
    df['deseasonalized'],
    'k', #colour black
    label='Deseasonalized')
ax.set_xlabel(
    'Year',
    fontstyle='italic') #labeling my x-axis & make it italic
ax.set_ylabel(
    r'$\mathit{CO_2}$ (ppm)',
    fontstyle='italic') #labeling my y-axis & make it italic
ax.set_xlim([2015,2025]) #interval for the x-axis
ax.set_ylim([395,430]) #interval for the x-axis
ax.set_title(
    r'Global average $\mathbf{CO_2}$ concentration',
    fontweight='bold') #labeling my title & make bold
ax.minorticks_on() #show the minor ticks
ax.legend(loc='upper left', labelcolor = 'linecolor') #show the legend with the labels given above in upper left corner
plt.grid(True)
```

```python
#generating the plot
fig, ax = plt.subplots(figsize=(12,5)) #set figure size to 15 inches by 10 inches
ax.plot( #plot time vs. average CO2 ppm
    df['decimal date'],
    df['average'],
    'r', #make line red
    alpha=0.5, #make line semi transparent
    label='Monthly Average')
ax.plot( #plot time vs. deseasonalized CO2 ppm on ax
    df['decimal date'],
    df['deseasonalized'],
    'k', #colour black
    label='Deseasonalized')
ax.set_xlabel(
    'Year',
    fontstyle='italic') #labeling my x-axis & make it italic
ax.set_ylabel(
    r'$\text{CO}_2$ concentration (ppm)',
    fontstyle='italic') #labeling my y-axis & make it italic
ax.set_title(
    r'Global average $\mathbf{CO_2}$ concentration',
    fontweight='bold') #labeling my title, adding subscript & make bold
ax.minorticks_on() #show the minor ticks
ax.legend(loc='lower right', labelcolor = 'linecolor') #show the legend with the labels given above in upper left corner
plt.grid(True)

#create container for the zoom
ax_ins = ax.inset_axes([0.15, 0.5, 0.3, 0.3])

# Plot data again
ax_ins.plot(df['decimal date'], df['average'], color='red')
ax_ins.plot(df['decimal date'], df['deseasonalized'], color='black')

# Define zoomed-in range (e.g., year 2010 to 2020)
x1, x2 = 2020, 2025
y1, y2 = 410, 430 # Manually set Y limits to show detail
ax_ins.set_xlim(x1, x2)
ax_ins.set_ylim(y1, y2)

#Hide ticks on the inset for a cleaner look
ax_ins.set_xticks([])
ax_ins.set_yticks([])

ax.indicate_inset_zoom(ax_ins, edgecolor="blue", linestyle='--')
plt.show()
```
